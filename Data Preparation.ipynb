```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df1["weight_diff_red-blue"]= df1["R_Weight_lbs"] - df1["B_Weight_lbs"] # Red less o more than Blue
```

```python
df1["height_diff_red-blue"]= df1["R_Height_cms"] - df1["B_Height_cms"] # Red less o more than Blue
```

```python
list(df1.columns)
```

###  We have to drop some unnecessary columns

```python
df1.drop(columns=[ "R_fighter","B_fighter", "Referee", "date","location","R_draw","B_draw","no_of_rounds","R_wins","B_wins",
                  "R_losses","B_losses","B_total_title_bouts","R_total_title_bouts"], inplace=True)
```

```python
df1.drop(columns=[ "R_total_rounds_fought","B_total_rounds_fought","R_total_time_fought(seconds)","B_total_time_fought(seconds)",
                  "R_longest_win_streak","B_longest_win_streak","R_current_lose_streak","B_current_lose_streak"], inplace=True)
```

```python
df1.drop(columns=['R_win_by_Decision_Majority', 'R_win_by_Decision_Split', 'R_win_by_Decision_Unanimous', 'R_win_by_KO/TKO',
                  'R_win_by_Submission','R_win_by_TKO_Doctor_Stoppage',"B_current_win_streak",'B_win_by_Decision_Majority',
                  'B_win_by_Decision_Split','B_win_by_Decision_Unanimous', 'B_win_by_KO/TKO', 'B_win_by_Submission',
                  'B_win_by_TKO_Doctor_Stoppage', ], inplace=True)
```

```python
df1.drop(columns=["R_current_win_streak" ], inplace=True)
```

```python
df1.drop(columns=['R_Stance','B_Stance'], inplace=True)
```

```python
df1.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[["Strawweight","Women's Strawweight", "Flyweight","Women's Flyweight", "Bantamweight","Women's Bantamweight","Featherweight","Women's Featherweight",
                                      "Lightweight","Welterweight","Middleweight","Light Heavyweight","Heavyweight","Open Weight","Catch Weight"]],dtype=np.int8)
df1['weight_class'] = encoder.fit_transform(df1['weight_class'].values.reshape(-1, 1))
```

```python
df1.title_bout=[0 if value == False else 1 for value in df1["title_bout"]]  # 0=False , 1=True
```

```python
df1.drop(df1[df1.Winner == "Draw"].index, inplace=True)
```

```python
df1["Winner"]= [0 if value == "Blue" else 1 for value in df1["Winner"]] # 1=Red, 0=Blue
```

```python
df1.to_csv('data/modeling_numeric.csv',index=False)
```
