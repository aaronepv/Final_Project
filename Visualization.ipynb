```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
df1 = pd.read_csv('data/final_dataset.csv')
```

```python
df2 = pd.read_csv('data/mens_fighters.csv')
```

```python
df3 = pd.read_csv('data/women_fighters.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

## We could observe that the age of the fighters is concentrated between 25 and 30 years old (Men Fighters)

```python
f,ax=plt.subplots(1,2,figsize=(10,8))
sns.distplot(df2['R_age'], ax=ax[0])

ax[0].set_title('R_age')
ax[0].set_ylabel('')
hist = sns.distplot(df2['B_age'],ax=ax[1])

ax[1].set_title('B_age')
plt.show();
```

## Most of the winners are between 28 and 30 years old (Men Fighters)

```python
f,ax=plt.subplots(1,2,figsize=(10,8))
df2[df2['Winner']=='Red']['R_age'].value_counts().plot.bar(ax=ax[0])

ax[0].set_title('R_age')
ax[0].set_ylabel('')
bar = df2[df2['Winner']=='Blue']['B_age'].value_counts().plot.bar(ax=ax[1])

ax[1].set_title('B_age')
plt.show();
```

## We could observe that the main location for the fight even are in USA

```python
df2['country'] = df2['location'].apply(lambda x : x.split(',')[-1])
```

```python
df2.country.value_counts().sort_values(ascending=True).plot(kind="barh");
```

## The most common class weight is Lightweight and  Welterweight (Men Fighters)

```python
plt.figure(figsize=(15,8))
values = df2['weight_class'].value_counts()
labels = values.index
sns.barplot(x=values,y=labels)
plt.title('UFC Weight Class')
plt.show()
```

## The most popular stance in the winners is orthodox (Men Fighters)

```python
f,ax=plt.subplots(1,2,figsize=(10,8))
df2[df2['Winner']=='Red']['R_Stance'].value_counts().plot.bar(ax=ax[0])

ax[0].set_title('R_Stance')
ax[0].set_ylabel('')
bar = df2[df2['Winner']=='Blue']['B_Stance'].value_counts().plot.bar(ax=ax[1])

ax[1].set_title('B_Stance')
plt.show();
```

## The most common class weight is Strawweight and  Bantamweight (Women Fighters)

```python
plt.figure(figsize=(15,8))
values = df3['weight_class'].value_counts()
labels = values.index
sns.barplot(x=values,y=labels)
plt.title('UFC Weight Class')
plt.show()
```

```python
df1["gender_bi"]= [0 if value == "M" else 1 for value in df1["gender"]] # 0=Male  , 1=Female
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.style.use('seaborn')

gender_.plot(kind="bar")
plt.xlabel('Gender')
plt.ylabel('Quantity')
plt.title('Quantity of Male and Female')
ticks = ["Male (4830)","Female (314)"]
ax.set_xticklabels(ticks)

plt.show()
```
